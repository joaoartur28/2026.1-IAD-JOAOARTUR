```python
import re
livro = input("Digite o nome do livro: ")

hand = open(livro)
for line in hand:
    line = line.rstrip()
    x = re.findall(r'[a-zA-Z0-9]\S*@\S*[a-zA-Z]', line)
    if len(x) > 0:
        print(x)
```
