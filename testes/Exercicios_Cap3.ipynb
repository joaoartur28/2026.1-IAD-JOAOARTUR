```python
# Reescreva o programa de cálculo de pagamento para pagar ao funcionário 1,5 vez o valor da hora normal por horas trabalhadas acima de 40.

horas_trabalhadas = float(input("Digite as horas trabalhadas:"))
taxa = float(input("Digite a taxa:"))

if horas_trabalhadas >= 40:
    hora_com_taxa = (horas_trabalhadas - 40) * (taxa * 1.5) + (taxa * 40)
    print(hora_com_taxa)
elif horas_trabalhadas < 40:
    hora_sem_taxa = horas_trabalhadas * taxa
    print(hora_sem_taxa)
```
