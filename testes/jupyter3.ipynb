```python
fname = input('coloque o nome do arquivo: ')
try:
    fhand = open(fname)
except:
    print('Documento não pode ser aberto', fname)
    exit()

counts = dict()  # 'c' minúsculo
for line in fhand:
    words = line.split()
    for word in words:  # corrigido para 'word'
        if word not in counts:  # 'c' minúsculo
            counts[word] = 1
        else:
            counts[word] = counts[word] + 1

print(counts)
```
