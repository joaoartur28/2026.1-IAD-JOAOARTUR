```python
# Substituindo termos em listas
t[1:3] = ['x','y']
print(t)
```

```python
# métodos para listas (adição de termos e etc)
t = ['a','b','c','d','e','f']
t.append("g")
print(t)
```

```python
# métodos para listas (adição de uma lista)
t = ['a','b','c','d','e','f']
t.extend(["g",'h'])
print(t)
```

```python
# sort ordena os elementos da lista
t = ['c','g','a','b','d','f','e']
t.sort()
print(t)
```

```python
# pop elimina o item mas ele retorna
t = ['a', 'b', 'c']
x = t.pop(1)
print(t)
print(x)
```

```python
# se não precisa do valor, então elimine com del
# você pode eliminar mais de um item utilizando [1:4] por exemplo
t = ['a', 'b', 'c']
del t[1]
print(t)
```

```python
# se sabe o elemento mas não sabe o indice, utilize o remove
t = ['a', 'b', 'c']
t.remove('b')
print(t)
```

```python
# Funções de listas
nums = [3, 4, 5, 10, 27, 46]
print(len(nums))  # Retorna o tamanho da lista
print(max(nums))  # Retorna o maior valor da lista
print(min(nums))  # Retorna o menor valor da lista
print(sum(nums))  # Retorna a soma dos valores da lista
print(sum(nums)/len(nums))  # Retorna a média dos valores da lista
```

```python
total = 0
count = 0
while True:
    inp = input('Digite um número:')
    if inp == "pronto":
        break
    value = float(inp)
    total = total + value
    count = count + 1
average = total / count
print('Média:', average)
```

```python
# Utiliza-se list para converter uma string em uma lista de caracteres
s = "spam"
t = list(s)
print(t)
```

```python
# Utiliza-se split para converter uma string em uma lista de palavras
s = "sentindo falta de você"
t = s.split()
print(t)
```

```python
# Join faz o caminho inverso de split
delimiter = " "
s = ['sentindo','falta','de','você']
t = delimiter.join(s)
print(t)
```
