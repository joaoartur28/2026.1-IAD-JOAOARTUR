```python
soma_total = 0
quantidade = 0

while True:
    entrada = input("Digite um número: ")
    
    # Condição de parada do loop
    if entrada.lower() == "pronto":
        break
        
    try:
        # Tenta converter a entrada para número decimal (float)
        numero = float(entrada)
        soma_total = soma_total + numero
        quantidade = quantidade + 1
    except ValueError:
        # Se a conversão falhar, captura o erro e avisa o usuário
        print("Entrada Inválida")
        continue

# Exibe os resultados apenas se algum número válido foi digitado
if quantidade > 0:
    media = soma_total / quantidade
    print(f"{soma_total} {quantidade} {media}")
else:
    print("Nenhum número válido foi digitado.")
```

```python
horas = float(input("Digite as Horas:"))
taxa = float(input("Digite a taxa:"))

try:
    if horas <= 40:
        pagamento = horas * taxa
    elif horas > 40:
        pagamento = 40 * taxa + (horas - 40) * taxa * 1.5
    print("O pagamento é:", pagamento)
except:
    print("Erro: Por favor, digite valores numéricos válidos.")
```
